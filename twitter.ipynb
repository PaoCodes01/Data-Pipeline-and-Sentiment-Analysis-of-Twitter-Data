```python
import pandas as pd
import sqlite3

# Load CSV
df = pd.read_csv("traffic_tweets_2922_anon.csv")

# Show CSV as a REAL table (not print)
print("CSV Preview:")
display(df.head(10))   # shows a nice HTML table
print()

# Connect to SQLite
conn = sqlite3.connect("traffic_tweets.db")

# Save to DB
df.to_sql("tweets", conn, if_exists="replace", index=False)

# Query from SQLite
query = "SELECT * FROM tweets LIMIT 20"
table = pd.read_sql(query, conn)

print("SQLite Table Preview:")
display(table)   # real table display

# Optional: schema
schema = pd.read_sql("PRAGMA table_info(tweets);", conn)
print("Table Schema:")
display(schema)

conn.close()
```
